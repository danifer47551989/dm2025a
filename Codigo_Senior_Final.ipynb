```r
drift_UVA <- function(campos_monetarios) {
  cat( "inicio drift_UVA()\n")

  dataset[tb_indices,
    on = c("foto_mes"),
    (campos_monetarios) := .SD * i.UVA,
    .SDcols = campos_monetarios
  ]

  cat( "fin drift_UVA()\n")
}
drift_dolar_oficial <- function(campos_monetarios) {
  cat( "inicio drift_dolar_oficial()\n")

  dataset[tb_indices,
    on = c("foto_mes"),
    (campos_monetarios) := .SD / i.dolar_oficial,
    .SDcols = campos_monetarios
  ]

  cat( "fin drift_dolar_oficial()\n")
}
drift_dolar_blue <- function(campos_monetarios) {
  cat( "inicio drift_dolar_blue()\n")

  dataset[tb_indices,
    on = c("foto_mes"),
    (campos_monetarios) := .SD / i.dolar_blue,
    .SDcols = campos_monetarios
  ]

  cat( "fin drift_dolar_blue()\n")
}
drift_deflacion <- function(campos_monetarios) {
  cat( "inicio drift_deflacion()\n")

  dataset[tb_indices,
    on = c("foto_mes"),
    (campos_monetarios) := .SD * i.IPC,
    .SDcols = campos_monetarios
  ]

  cat( "fin drift_deflacion()\n")
}
drift_rank_simple <- function(campos_drift) {

  cat( "inicio drift_rank_simple()\n")
  for (campo in campos_drift)
  {
    cat(campo, " ")
    dataset[, paste0(campo, "_rank") :=
      (frank(get(campo), ties.method = "random") - 1) / (.N - 1), by = list(foto_mes)]
    dataset[, (campo) := NULL]
  }
  cat( "fin drift_rank_simple()\n")
}
# El cero se transforma en cero
# los positivos se rankean por su lado
# los negativos se rankean por su lado

drift_rank_cero_fijo <- function(campos_drift) {

  cat( "inicio drift_rank_cero_fijo()\n")
  for (campo in campos_drift)
  {
    cat(campo, " ")
    dataset[get(campo) == 0, paste0(campo, "_rank") := 0]
    dataset[get(campo) > 0, paste0(campo, "_rank") :=
      frank(get(campo), ties.method = "random") / .N, by = list(foto_mes)]

    dataset[get(campo) < 0, paste0(campo, "_rank") :=
      -frank(-get(campo), ties.method = "random") / .N, by = list(foto_mes)]
    dataset[, (campo) := NULL]
  }
  cat("\n")
  cat( "fin drift_rank_cero_fijo()\n")
}
drift_estandarizar <- function(campos_drift) {

  cat( "inicio drift_estandarizar()\n")
  for (campo in campos_drift)
  {
    cat(campo, " ")
    dataset[, paste0(campo, "_normal") :=
      (get(campo) -mean(campo, na.rm=TRUE)) / sd(get(campo), na.rm=TRUE),
      by = list(foto_mes)]

    dataset[, (campo) := NULL]
  }
  cat( "fin drift_estandarizar()\n")
}
```

#### 9.7.1.4  FEhist Feature Engineering historico

El Fature Engineering Histórico es la etapa que más aporta a la ganancia final, ya que enriquece cada registro del dataset con su historia.

Para cada campo del dataset original (*)
se crean lo siguientes campos de a partir de la historia
* lag1  lags de orden 1
* delta1  =  valor actual - lag1
* lag2  lags de orden 2
* delta2  = valor actual - lag2


(*) Excepto para los campos  <numero_de_cliente,  foto_mes,  clase_ternaria>

```r
# se calculan para los 6 meses previos el minimo, maximo y
#  tendencia calculada con cuadrados minimos
# la formula de calculo de la tendencia puede verse en
#  https://stats.libretexts.org/Bookshelves/Introductory_Statistics/Book%3A_Introductory_Statistics_(Shafer_and_Zhang)/10%3A_Correlation_and_Regression/10.04%3A_The_Least_Squares_Regression_Line
# para la maxíma velocidad esta funcion esta escrita en lenguaje C,
# y no en la porqueria de R o Python

cppFunction("NumericVector fhistC(NumericVector pcolumna, IntegerVector pdesde )
{
  /* Aqui se cargan los valores para la regresion */
  double  x[100] ;
  double  y[100] ;

  int n = pcolumna.size();
  NumericVector out( 5*n );

  for(int i = 0; i < n; i++)
  {
    //lag
    if( pdesde[i]-1 < i )  out[ i + 4*n ]  =  pcolumna[i-1] ;
    else                   out[ i + 4*n ]  =  NA_REAL ;


    int  libre    = 0 ;
    int  xvalor   = 1 ;

    for( int j= pdesde[i]-1;  j<=i; j++ )
    {
       double a = pcolumna[j] ;

       if( !R_IsNA( a ) )
       {
          y[ libre ]= a ;
          x[ libre ]= xvalor ;
          libre++ ;
       }

       xvalor++ ;
    }

    /* Si hay al menos dos valores */
    if( libre > 1 )
    {
      double  xsum  = x[0] ;
      double  ysum  = y[0] ;
      double  xysum = xsum * ysum ;
      double  xxsum = xsum * xsum ;
      double  vmin  = y[0] ;
      double  vmax  = y[0] ;

      for( int h=1; h<libre; h++)
      {
        xsum  += x[h] ;
        ysum  += y[h] ;
        xysum += x[h]*y[h] ;
        xxsum += x[h]*x[h] ;

        if( y[h] < vmin )  vmin = y[h] ;
        if( y[h] > vmax )  vmax = y[h] ;
      }

      out[ i ]  =  (libre*xysum - xsum*ysum)/(libre*xxsum -xsum*xsum) ;
      out[ i + n ]    =  vmin ;
      out[ i + 2*n ]  =  vmax ;
      out[ i + 3*n ]  =  ysum / libre ;
    }
    else
    {
      out[ i       ]  =  NA_REAL ;
      out[ i + n   ]  =  NA_REAL ;
      out[ i + 2*n ]  =  NA_REAL ;
      out[ i + 3*n ]  =  NA_REAL ;
    }
  }

  return  out;
}")
```

```r
# calcula la tendencia de las variables cols de los ultimos 6 meses
# la tendencia es la pendiente de la recta que ajusta por cuadrados minimos
# La funcionalidad de ratioavg es autoria de  Daiana Sparta,  UAustral  2021

TendenciaYmuchomas <- function(
    dataset, cols, ventana = 6, tendencia = TRUE,
    minimo = TRUE, maximo = TRUE, promedio = TRUE,
    ratioavg = FALSE, ratiomax = FALSE) {
  gc(verbose= FALSE)
  # Esta es la cantidad de meses que utilizo para la historia
  ventana_regresion <- ventana

  last <- nrow(dataset)

  # creo el vector_desde que indica cada ventana
  # de esta forma se acelera el procesamiento ya que lo hago una sola vez
  vector_ids <- dataset[ , numero_de_cliente ]

  vector_desde <- seq(
    -ventana_regresion + 2,
    nrow(dataset) - ventana_regresion + 1
  )

  vector_desde[1:ventana_regresion] <- 1

  for (i in 2:last) {
    if (vector_ids[i - 1] != vector_ids[i]) {
      vector_desde[i] <- i
    }
  }
  for (i in 2:last) {
    if (vector_desde[i] < vector_desde[i - 1]) {
      vector_desde[i] <- vector_desde[i - 1]
    }
  }

  for (campo in cols) {
    nueva_col <- fhistC(dataset[, get(campo)], vector_desde)

    if (tendencia) {
      dataset[, paste0(campo, "_tend", ventana) :=
        nueva_col[(0 * last + 1):(1 * last)]]
    }

    if (minimo) {
      dataset[, paste0(campo, "_min", ventana) :=
        nueva_col[(1 * last + 1):(2 * last)]]
    }

    if (maximo) {
      dataset[, paste0(campo, "_max", ventana) :=
        nueva_col[(2 * last + 1):(3 * last)]]
    }

    if (promedio) {
      dataset[, paste0(campo, "_avg", ventana) :=
        nueva_col[(3 * last + 1):(4 * last)]]
    }

    if (ratioavg) {
      dataset[, paste0(campo, "_ratioavg", ventana) :=
        get(campo) / nueva_col[(3 * last + 1):(4 * last)]]
    }

    if (ratiomax) {
      dataset[, paste0(campo, "_ratiomax", ventana) :=
        get(campo) / nueva_col[(2 * last + 1):(3 * last)]]
    }
  }
}
```

```r
AgregaVarRandomForest <- function() {

  cat( "inicio AgregaVarRandomForest()\n")
  gc(verbose= FALSE)
  dataset[, clase01 := 0L ]
  dataset[ clase_ternaria %in% PARAM$FE_rf$train$clase01_valor1,
      clase01 := 1L ]

  campos_buenos <- setdiff(
    colnames(dataset),
    c( "clase_ternaria", "clase01")
  )

  dataset[, entrenamiento :=
    as.integer( foto_mes %in% PARAM$FE_rf$train$training )]

  dtrain <- lgb.Dataset(
    data = data.matrix(dataset[entrenamiento == TRUE, campos_buenos, with = FALSE]),
    label = dataset[entrenamiento == TRUE, clase01],
    free_raw_data = FALSE
  )

  modelo <- lgb.train(
     data = dtrain,
     param = PARAM$FE_rf$lgb_param,
     verbose = -100
  )

  cat( "Fin construccion RandomForest\n" )
  # grabo el modelo, achivo .model
  lgb.save(modelo, file="modelo.model" )

  qarbolitos <- copy(PARAM$FE_rf$lgb_param$num_iterations)

  periodos <- dataset[ , unique( foto_mes ) ]

  for( periodo in  periodos )
  {
    cat( "periodo = ", periodo, "\n" )
    datamatrix <- data.matrix(dataset[ foto_mes== periodo, campos_buenos, with = FALSE])

    cat( "Inicio prediccion\n" )
    prediccion <- predict(
        modelo,
        datamatrix,
        type = "leaf"
    )
    cat( "Fin prediccion\n" )

    for( arbolito in 1:qarbolitos )
    {
       cat( arbolito, " " )
       hojas_arbol <- unique(prediccion[ , arbolito])

       for (pos in 1:length(hojas_arbol)) {
         # el numero de nodo de la hoja, estan salteados
         nodo_id <- hojas_arbol[pos]
         dataset[ foto_mes== periodo, paste0(
            "rf_", sprintf("%03d", arbolito),
             "_", sprintf("%03d", nodo_id)
          ) :=  as.integer( nodo_id == prediccion[ , arbolito]) ]

       }

       rm( hojas_arbol )
    }
    cat( "\n" )

    rm( prediccion )
    rm( datamatrix )
    gc(verbose= FALSE)
  }

  gc(verbose= FALSE)

  # borro clase01 , no debe ensuciar el dataset
  dataset[ , clase01 := NULL ]

}
```

```r
#GA
# ==================== 6.3.1.6-ter — GA2: GENERACIÓN de features (inspirado JAR, optimizado) ====================

if (!requireNamespace("GA", quietly = TRUE)) install.packages("GA")
if (!requireNamespace("lightgbm", quietly = TRUE)) install.packages("lightgbm")
if (!requireNamespace("data.table", quietly = TRUE)) install.packages("data.table")
suppressPackageStartupMessages({
  library(GA); library(lightgbm); library(data.table)
})

# ---------- configuración ----------
GA2_SUFFIXES      <- c("", "_lag1", "_delta1", "_lag2", "_delta2")
GA2_FGEN          <- 10L
GA2_MAX_COLS      <- 512L
GA2_BASE_LIMIT    <- 256L
GA2_POPSIZE       <- 100L
GA2_MAXITER       <- 60L
GA2_RUN_NOIMPROVE <- 15L
GA2_EPS           <- 1e-6

PARALLEL_CORES <- max(1L, parallel::detectCores(logical = TRUE))
set.seed(PARAM$semilla_primigenia)

# ---------- pool base ----------
if (exists("campos_buenos")) {
  base_pool <- intersect(campos_buenos, colnames(dataset))
} else {
  base_pool <- setdiff(colnames(dataset),
                       c("numero_de_cliente","foto_mes","clase_ternaria","clase01","azar"))
}
stopifnot(length(base_pool) > 0)

# ---------- train/valid ----------
val_meses_disponibles <- sort(unique(dataset$foto_mes))
GA_validate_meses <- tryCatch(PARAM$trainingstrategy$validate, error=function(e) NULL)
if (is.null(GA_validate_meses)) GA_validate_meses <- tail(val_meses_disponibles, 1)

idx_valid_GA <- which(dataset$foto_mes %in% GA_validate_meses)
idx_train_GA <- which(!(dataset$foto_mes %in% GA_validate_meses))

y_all <- as.integer(dataset$clase_ternaria %in% c("BAJA+1","BAJA+2"))

# subsample (igual que tenías)
GA_use_subsample <- TRUE
GA_pos_max <- 100000L; GA_neg_max <- 100000L
if (GA_use_subsample) {
  pos_idx <- idx_train_GA[y_all[idx_train_GA] == 1]
  neg_idx <- idx_train_GA[y_all[idx_train_GA] == 0]
  set.seed(PARAM$semilla_primigenia)
  pos_take <- if (length(pos_idx) > GA_pos_max) sample(pos_idx, GA_pos_max) else pos_idx
  neg_take <- if (length(neg_idx) > GA_neg_max) sample(neg_idx, GA_neg_max) else neg_idx
  idx_train_GA <- c(pos_take, neg_take)
}
cat("GA2 idx -> train:", length(idx_train_GA), " valid:", length(idx_valid_GA), "\n")
stopifnot(length(idx_train_GA) > 0, length(idx_valid_GA) > 0)

# ---------- baseline (top |cor|) ----------
cor_sc <- suppressWarnings(sapply(base_pool, function(v) {
  x <- dataset[[v]][idx_train_GA]
  if (!is.numeric(x)) return(0)
  if (all(!is.finite(x))) return(0)
  tryCatch(abs(cor(x, y_all[idx_train_GA], use="pairwise.complete.obs")), error=function(e) 0)
}))
base_top <- names(sort(cor_sc, decreasing=TRUE))[seq_len(min(GA2_BASE_LIMIT, length(cor_sc)))]
Xtr_base <- data.matrix(dataset[idx_train_GA, base_top, with=FALSE])
Xva_base <- data.matrix(dataset[idx_valid_GA, base_top, with=FALSE])
ytr <- y_all[idx_train_GA]; yva <- y_all[idx_valid_GA]
N_BASE <- ncol(Xtr_base)

# ---------- helpers ultra-ligeros ----------
# existencia rápida de columna con sufijo
.has_with_suffix <- function(name, suffix) {
  col <- if (suffix == "") name else paste0(name, suffix)
  col %in% colnames(dataset)
}
get_with_suffix <- function(name, suffix, rows) {
  col <- if (suffix=="") name else paste0(name, suffix)
  if (!(col %in% colnames(dataset))) return(NULL)
  dataset[[col]][rows]
}
signed_power <- function(x, a) {
  sx <- sign(x); ax <- abs(x)
  sx * (ax + GA2_EPS)^a
}
OPS2 <- c("sum","diff","prod","ratio","share")

decode_gen_features2 <- function(x_tail) {
  out <- vector("list", GA2_FGEN)
  pos <- 1L
  for (k in seq_len(GA2_FGEN)) {
    ia <- 1 + floor(x_tail[pos + 0] * length(base_pool)); ia <- min(ia, length(base_pool))
    sa <- 1 + floor(x_tail[pos + 1] * length(GA2_SUFFIXES)); sa <- min(sa, length(GA2_SUFFIXES))
    aa <- 0.25 + x_tail[pos + 2] * (2.0 - 0.25)

    ib <- 1 + floor(x_tail[pos + 3] * length(base_pool)); ib <- min(ib, length(base_pool))
    sb <- 1 + floor(x_tail[pos + 4] * length(GA2_SUFFIXES)); sb <- min(sb, length(GA2_SUFFIXES))
    ab <- 0.25 + x_tail[pos + 5] * (2.0 - 0.25)

    io <- 1 + floor(x_tail[pos + 6] * length(OPS2)); io <- min(io, length(OPS2))
    op <- OPS2[io]

    if (ia == ib) ib <- ((ia + 1L) %% length(base_pool)) + 1L

    out[[k]] <- list(
      a_name = base_pool[ia], a_suf = GA2_SUFFIXES[sa], a_alpha = aa,
      b_name = base_pool[ib], b_suf = GA2_SUFFIXES[sb], b_alpha = ab,
      op = op
    )
    pos <- pos + 7L
  }
  out
}

.pcode <- function(a) sprintf("p%02d", round(100*a))
.mk_name2 <- function(def) {
  paste0("gaj_", def$op, "_",
         def$a_name, def$a_suf, "_", .pcode(def$a_alpha), "__",
         def$b_name, def$b_suf, "_", .pcode(def$b_alpha))
}

materialize_one2 <- function(def, idx) {
  # skip temprano si faltan columnas
  if (!.has_with_suffix(def$a_name, def$a_suf) || !.has_with_suffix(def$b_name, def$b_suf))
    return(NULL)
  a <- get_with_suffix(def$a_name, def$a_suf, idx)
  b <- get_with_suffix(def$b_name, def$b_suf, idx)
  if (is.null(a) || is.null(b)) return(NULL)
  if (!any(is.finite(a)) || !any(is.finite(b))) return(NULL)

  a <- signed_power(a, def$a_alpha)
  b <- signed_power(b, def$b_alpha)

  if (def$op == "sum")   return(a + b)
  if (def$op == "diff")  return(a - b)
  if (def$op == "prod")  return(a * b)
  if (def$op == "ratio") return(a / (abs(b) + GA2_EPS))
  if (def$op == "share") return(a / (abs(a) + abs(b) + GA2_EPS))
  NULL
}

# byte-compile las hot paths si está disponible
if (requireNamespace("compiler", quietly = TRUE)) {
  decode_gen_features2 <- compiler::cmpfun(decode_gen_features2)
  materialize_one2     <- compiler::cmpfun(materialize_one2)
}

# ---------- LGBM params (1 hilo) ----------
get_base_params_for_GA2 <- function() {
  list(
    objective               = "binary",
    metric                  = "auc",
    first_metric_only       = TRUE,
    boost_from_average      = TRUE,
    feature_pre_filter      = FALSE,
    verbosity               = -100,
    force_row_wise          = TRUE,
    seed                    = PARAM$semilla_primigenia,
    max_bin                 = 31,
    num_iterations          = 768,
    early_stopping_rounds   = 200,
    min_sum_hessian_in_leaf = 10.0,
    bagging_fraction        = 0.8,
    bagging_freq            = 1,
    learning_rate           = 0.05,
    feature_fraction        = 0.7,
    min_data_in_leaf        = 290,
    num_leaves              = 64,
    num_threads             = 1L
  )
}

# ---------- fitness con caché (y sin cbind repetidos) ----------
.FIT2_CACHE <- new.env(parent=emptyenv())
penalize_complexity2 <- function(auc, p_new) auc - 8e-4 * p_new

ga2_fitness <- function(x) {
  key <- paste(round(x,4), collapse="|")
  if (exists(key, envir=.FIT2_CACHE)) return(get(key, envir=.FIT2_CACHE))

  defs <- decode_gen_features2(x)

  gen_tr <- vector("list", length(defs))
  gen_va <- vector("list", length(defs))
  gen_nm <- character(length(defs))
  added  <- 0L

  for (i in seq_along(defs)) {
    d <- defs[[i]]
    vtr <- materialize_one2(d, idx_train_GA)
    if (is.null(vtr)) next
    if (!any(is.finite(vtr))) next
    vv <- var(vtr[is.finite(vtr)], na.rm = TRUE)
    if (!is.finite(vv) || is.na(vv) || vv == 0) next

    vva <- materialize_one2(d, idx_valid_GA)
    if (is.null(vva)) next

    added <- added + 1L
    nm <- .mk_name2(d)
    gen_tr[[added]] <- as.numeric(vtr)
    gen_va[[added]] <- as.numeric(vva)
    gen_nm[added]   <- nm
  }

  if (added == 0L) { assign(key, -1e6, envir=.FIT2_CACHE); return(-1e6) }

  # recorta listas al tamaño real
  if (added < length(gen_tr)) {
    gen_tr <- gen_tr[seq_len(added)]
    gen_va <- gen_va[seq_len(added)]
    gen_nm <- gen_nm[seq_len(added)]
  }

  # ---- cap inteligente: siempre conservo baseline y selecciono top de generadas por |cor|
  if (N_BASE + added <= GA2_MAX_COLS) {
    Xtr <- if (added > 0L) cbind(Xtr_base, do.call(cbind, gen_tr)) else Xtr_base
    Xva <- if (added > 0L) cbind(Xva_base, do.call(cbind, gen_va)) else Xva_base
    colnames(Xtr) <- c(colnames(Xtr_base), gen_nm)
    colnames(Xva) <- c(colnames(Xva_base), gen_nm)
  } else {
    slots_add <- max(0L, GA2_MAX_COLS - N_BASE)
    if (slots_add == 0L) {
      Xtr <- Xtr_base; Xva <- Xva_base
    } else {
      # correlación SOLO para las generadas (mucho más barato)
      sc <- sapply(gen_tr, function(v) {
        v <- as.numeric(v)
        suppressWarnings(tryCatch(abs(cor(v, ytr, use="pairwise.complete.obs")), error=function(e) 0))
      })
      take <- order(sc, decreasing = TRUE)[seq_len(slots_add)]
      Xtr <- cbind(Xtr_base, do.call(cbind, gen_tr[take]))
      Xva <- cbind(Xva_base, do.call(cbind, gen_va[take]))
      colnames(Xtr) <- c(colnames(Xtr_base), gen_nm[take])
      colnames(Xva) <- c(colnames(Xva_base), gen_nm[take])
    }
  }

  param <- get_base_params_for_GA2()

  dtr <- try(lgb.Dataset(data = Xtr, label = ytr, free_raw_data = TRUE), silent = TRUE)
  if (inherits(dtr, "try-error")) { assign(key, -1e6, envir=.FIT2_CACHE); return(-1e6) }
  dva <- try(lgb.Dataset(data = Xva, label = yva, free_raw_data = TRUE), silent = TRUE)
  if (inherits(dva, "try-error")) { rm(dtr); assign(key, -1e6, envir=.FIT2_CACHE); return(-1e6) }

  modelo <- try(lgb.train(data = dtr, valids = list(valid = dva),
                          eval = "auc", param = param, verbose = -100), silent = TRUE)
  if (inherits(modelo, "try-error")) { rm(dtr, dva); gc(FALSE); assign(key, -1e6, envir=.FIT2_CACHE); return(-1e6) }

  AUC <- tryCatch({
    if (!is.null(modelo$best_score)) as.numeric(modelo$best_score) else
      as.numeric(modelo$record_evals$valid$auc$eval[[modelo$best_iter]])
  }, error=function(e) NA_real_)

  rm(modelo, dtr, dva); gc(FALSE)

  if (!is.finite(AUC)) { assign(key, -1e6, envir=.FIT2_CACHE); return(-1e6) }

  val <- penalize_complexity2(AUC, p_new = added)
  assign(key, val, envir=.FIT2_CACHE)
  val
}

# ---------- checkpoint + monitor seguros ----------
GA2_CKPT_FILE <- file.path(getwd(), "GA2_checkpoint.rds")
ga2_monitor <- function(obj){
  best_fit <- suppressWarnings(max(obj@fitness, na.rm=TRUE))
  if (!is.finite(best_fit)) best_fit <- NA_real_
  cat(sprintf("GA2 Gen %02d | Best AUC*=%s | pop=%d\n",
              obj@iter,
              ifelse(is.na(best_fit), "NA", sprintf("%.6f", best_fit)),
              obj@popSize))
  best_sol <- NULL
  if (is.matrix(obj@solution) && nrow(obj@solution) >= 1) best_sol <- obj@solution[1, , drop = TRUE]
  if (is.numeric(best_sol) && length(best_sol) == (GA2_FGEN * 7L) && all(is.finite(best_sol))) {
    state <- list(iter = obj@iter, best_fit = best_fit, best_sol = best_sol)
    try(saveRDS(state, GA2_CKPT_FILE), silent = TRUE)
  }
  invisible(TRUE)
}

suggestions_ckpt <- NULL
if (file.exists(GA2_CKPT_FILE)) {
  ck <- try(readRDS(GA2_CKPT_FILE), silent = TRUE)
  if (!inherits(ck, "try-error") && is.numeric(ck$best_sol)) {
    if (length(ck$best_sol) == (GA2_FGEN * 7L) && all(is.finite(ck$best_sol))) {
      suggestions_ckpt <- matrix(ck$best_sol, nrow = 1)
    }
  }
}

# limitar threads de data.table/BLAS durante el GA
old_dt_threads <- data.table::getDTthreads()
data.table::setDTthreads(1)
if (requireNamespace("RhpcBLASctl", quietly = TRUE)) {
  RhpcBLASctl::blas_set_num_threads(1)
  RhpcBLASctl::omp_set_num_threads(1)
}

# ---------- ejecutar GA2 ----------
nvars2 <- GA2_FGEN * 7L
GA2 <- ga(
  type       = "real-valued",
  fitness    = ga2_fitness,
  lower      = rep(0, nvars2),
  upper      = rep(1, nvars2),
  popSize    = GA2_POPSIZE,
  maxiter    = GA2_MAXITER,
  run        = GA2_RUN_NOIMPROVE,
  pmutation  = 0.2,
  pcrossover = 0.8,
  elitism    = 2,
  parallel   = PARALLEL_CORES,
  suggestions= suggestions_ckpt,
  monitor    = ga2_monitor
)
print(summary(GA2))

data.table::setDTthreads(old_dt_threads)

# ---------- materializar MEJOR individuo ----------
best2 <- NULL
if (is.matrix(GA2@solution) && nrow(GA2@solution) >= 1) best2 <- GA2@solution[1, , drop = TRUE]
if (is.null(best2) || length(best2) != (GA2_FGEN * 7L)) {
  idx_best <- suppressWarnings(which.max(GA2@fitness))
  if (length(idx_best) == 1 && is.matrix(GA2@population)) best2 <- GA2@population[idx_best, ]
}
stopifnot(is.numeric(best2), length(best2) == (GA2_FGEN * 7L))

defs_best2 <- decode_gen_features2(best2)

new_names2 <- character(0)
for (d in defs_best2) {
  nm <- .mk_name2(d)
  if (!nm %in% colnames(dataset)) {
    v <- materialize_one2(d, seq_len(nrow(dataset)))
    if (!is.null(v)) dataset[[nm]] <- v
  }
  new_names2 <- c(new_names2, nm)
}

PARAM$out$ga2 <- list(
  new_features = new_names2,
  auc_star     = suppressWarnings(max(GA2@fitness, na.rm=TRUE))
)

if (exists("campos_buenos")) {
  campos_buenos <- unique(c(campos_buenos, new_names2))
}

cat("GA2: nuevas features generadas =", length(new_names2), "\n")
cat("Dataset post-GA2 -> ncol =", ncol(dataset), "\n")
try({
  fwrite(data.table(new_feature = new_names2), file = "ga2_new_features.txt", sep = "\t")
  fwrite(data.table(colname = colnames(dataset)), file = "dataset_cols_after_ga2.txt", sep = "\t")
}, silent = TRUE)

# ==================== fin 6.3.1.6-ter (optimizado) ====================
```

#### 9.7.1.6  FEhist Reduccion dimensionalidad con canaritos

Esta etapa solo se mostrará a la *modalidad Anlista Sr*

El objetivo de esta etapa NO es mejorar el modelo predictivo

El objetivo es eliminar campos poco importantes para hacer espacio a nuevos campos, debido a las restricciones de memoria RAM.

```r
VPOS_CORTE <- c()

fganancia_lgbm_meseta <- function(probs, datos) {
  vlabels <- get_field(datos, "label")
  vpesos <- get_field(datos, "weight")

  tbl <- as.data.table(list(
    "prob" = probs,
    "gan" = ifelse(vlabels == 1 & vpesos > 1, PARAM$CN$train$gan1, PARAM$CN$train$gan0)
  ))

  setorder(tbl, -prob)
  tbl[, posicion := .I]
  tbl[, gan_acum := cumsum(gan)]
  setorder(tbl, -gan_acum) # voy por la meseta

  gan <- mean(tbl[1:500, gan_acum]) # meseta de tamaño 500

  pos_meseta <- tbl[1:500, median(posicion)]
  VPOS_CORTE <<- c(VPOS_CORTE, pos_meseta)

  return(list(
    "name" = "ganancia",
    "value" = gan,
    "higher_better" = TRUE
  ))
}
```

```r
# Elimina del dataset las variables que estan por debajo
#  de la capa geologica de canaritos
# se llama varias veces, luego de agregar muchas variables nuevas,
#  para ir reduciendo la cantidad de variables
# y así hacer lugar a nuevas variables importantes

GVEZ <- 1

campitos <- c( "numero_de_cliente", "foto_mes", "clase_ternaria" )

CanaritosAsesinos <- function(
  canaritos_ratio,
  canaritos_desvios,
  canaritos_semilla) {

  cat( "inicio CanaritosAsesinos()\n")
  gc(verbose= FALSE)
  dataset[, clase01 := 0L ]
  dataset[ clase_ternaria %in% PARAM$CN$train$clase01_valor1,
      clase01 := 1L ]

  set.seed(canaritos_semilla, kind = "L'Ecuyer-CMRG")
  for (i in 1:(ncol(dataset) * canaritos_ratio)) {
    dataset[, paste0("canarito", i) := runif(nrow(dataset))]
  }

  campos_buenos <- setdiff(
    colnames(dataset),
    c( campitos, "clase01")
  )

  azar <- runif(nrow(dataset))

  dataset[, entrenamiento :=
    as.integer( foto_mes %in% PARAM$CN$train$training &
      (clase01 == 1 | azar < PARAM$CN$train$undersampling))]

  dtrain <- lgb.Dataset(
    data = data.matrix(dataset[entrenamiento == TRUE, campos_buenos, with = FALSE]),
    label = dataset[entrenamiento == TRUE, clase01],
    weight = dataset[
      entrenamiento == TRUE,
      ifelse(clase_ternaria %in% PARAM$CN$train$positivos, 1.0000001, 1.0)
    ],
    free_raw_data = FALSE
  )

  dvalid <- lgb.Dataset(
    data = data.matrix(dataset[foto_mes %in% PARAM$CN$train$validation, campos_buenos, with = FALSE]),
    label = dataset[foto_mes %in% PARAM$CN$train$validation, clase01],
    weight = dataset[
      foto_mes %in% PARAM$CN$train$validation,
      ifelse( clase_ternaria %in% PARAM$CN$train$positivos, 1.0000001, 1.0)
    ],
    free_raw_data = FALSE
  )


  param <- list(
    objective = "binary",
    metric = "custom",
    first_metric_only = TRUE,
    boost_from_average = TRUE,
    feature_pre_filter = FALSE,
    verbosity = -100,
    seed = canaritos_semilla,
    max_depth = -1, # -1 significa no limitar,  por ahora lo dejo fijo
    min_gain_to_split = 0.0, # por ahora, lo dejo fijo
    lambda_l1 = 0.0, # por ahora, lo dejo fijo
    lambda_l2 = 0.0, # por ahora, lo dejo fijo
    max_bin = 31, # por ahora, lo dejo fijo
    num_iterations = 9999, # un numero grande, lo limita early_stopping_rounds
    force_row_wise = TRUE, # para que los alumnos no se atemoricen con  warning
    learning_rate = 0.065,
    feature_fraction = 1.0, # lo seteo en 1
    min_data_in_leaf = 260,
    num_leaves = 60,
    early_stopping_rounds = 200,
    num_threads = 1
  )

  set.seed(canaritos_semilla, kind = "L'Ecuyer-CMRG")
  modelo <- lgb.train(
    data = dtrain,
    valids = list(valid = dvalid),
    eval = fganancia_lgbm_meseta,
    param = param,
    verbose = -100
  )

  tb_importancia <- lgb.importance(model = modelo)
  tb_importancia[, pos := .I]

  fwrite(tb_importancia,
    file = paste0("impo_", GVEZ, ".txt"),
    sep = "\t"
  )

  GVEZ <<- GVEZ + 1

  umbral <- tb_importancia[
    Feature %like% "canarito",
    median(pos) + canaritos_desvios * sd(pos)
  ] # Atencion corto en la mediana mas desvios!!

  col_utiles <- tb_importancia[
    pos < umbral & !(Feature %like% "canarito"),
    Feature
  ]

  col_utiles <- unique(c(
    col_utiles,
    c(campitos, "mes")
  ))

  col_inutiles <- setdiff(colnames(dataset), col_utiles)

  dataset[, (col_inutiles) := NULL]

  cat( "fin CanaritosAsesinos()\n")

  return( tb_importancia )
}
```

Definición de la Bayesian Optimization
<br> Si se desea optimizar un hiperparámetro que esta como fijo, debe QUITARSE de param_fijos y agregarse a PARAM$hipeparametertuning$hs

```r
# 40 es el minimo razonable,  un buen analista Sr aumentara este valor
PARAM$hipeparametertuning$num_interations <- 80

# parametros fijos del LightGBM
PARAM$lgbm$param_fijos <- list(
  objective= "binary",
  metric= "auc",
  first_metric_only= TRUE,
  boost_from_average= TRUE,
  feature_pre_filter= FALSE,
  verbosity= -100,
  force_row_wise= TRUE, # para evitar warning
  seed= PARAM$semilla_primigenia,
  extra_trees = FALSE,

  max_depth = -1L, # -1 significa no limitar,  por ahora lo dejo fijo
  min_gain_to_split = 0.08, # min_gain_to_split >= 0.0
  min_sum_hessian_in_leaf = 0.005, #  min_sum_hessian_in_leaf >= 0.0
  lambda_l1 = 0.0, # lambda_l1 >= 0.0
  lambda_l2 = 100.0, # lambda_l2 >= 0.0

  bagging_fraction = 0.8, # 0.0 < bagging_fraction <= 1.0
  pos_bagging_fraction = 1.0, # 0.0 < pos_bagging_fraction <= 1.0
  neg_bagging_fraction = 0.8, # 0.0 < neg_bagging_fraction <= 1.0
  is_unbalance = FALSE, #
  scale_pos_weight = 1.0, # scale_pos_weight > 0.0

  drop_rate = 0.1, # 0.0 < neg_bagging_fraction <= 1.0
  max_drop = 50, # <=0 means no limit
  skip_drop = 0.5, # 0.0 <= skip_drop <= 1.0

  max_bin= 31,
  num_iterations= 2048,  # valor grande, lo limita early_stopping_rounds
  early_stopping_rounds= 200
)

```

Nuevos  pseudo hiperparámetros

*   leaf_size (0.0,  0.5]  es el   min_data_in_leaf / nrow(dataset)
*   coverage (0.0, 1.0]  se cumple   num_leaves =   coverage * (nrow(dataset)/min_data_in_leaf)



```r
# Notar que se usan los pseudo hiperparametros "ratio"  leaf_size  y coverage
#  que luego terminan en  min_data_in_leaf y  num_leaves
#  de forma que sea invariante al tamaño del dataset
# donde definís la par.set de MBO
PARAM$hipeparametertuning$hs <- makeParamSet(
  makeNumericParam("learning_rate", lower=0.01, upper=0.5),
  makeNumericParam("feature_fraction", lower=0.4, upper=0.9),
  makeIntegerParam("num_leaves", lower=32, upper=128),
  makeNumericParam("coverage", lower = 0.04, upper = 1.5), # nuevo
  makeNumericParam("leaf_size", lower = 0.001, upper = 0.80), # nuevo
  ## NUEVO según recomendaciones:
  makeNumericParam("lambda_l1",        lower=0,   upper=300),
  makeNumericParam("lambda_l2",        lower=0,   upper=1000),
  makeNumericParam("min_gain_to_split",lower=0.0, upper=20.0)
)
```

Función "señora caja negra"  que es llamada para verificar la realidad por la Bayesian Optimization

```r
# Toggle para BO
PARAM$bo <- modifyList(
  list(eval_multi_seed = TRUE, agg = "mean"),  # agg: "mean" | "median" | "p25" | "p75"
  if (exists("PARAM") && !is.null(PARAM$bo)) PARAM$bo else list()
)

EstimarGanancia_AUC_lightgbm <- function(x){

  # mezcla hparams fijos + x
  p_base <- modifyList(PARAM$lgbm$param_fijos, x)

  # mapea leaf_size/coverage -> min_data_in_leaf/num_leaves si los usás en el search space
  mapear_hparams <- function(p){
    if (!is.null(p$leaf_size))
      p$min_data_in_leaf <- max(1L, round(nrow(dtrain) * p$leaf_size))
    if (!is.null(p$coverage)) {
      if (is.null(p$min_data_in_leaf))
        p$min_data_in_leaf <- max(1L, round(nrow(dtrain) * 0.01))
      p$num_leaves <- min(131072L, max(8L, round(p$coverage * nrow(dtrain) / p$min_data_in_leaf)))
    }
    p
  }

  seeds <- as.integer(PARAM$semillas)

  if (isTRUE(PARAM$bo$eval_multi_seed) && length(seeds) > 1L) {
    # evalúo el MISMO x en varias semillas, en paralelo
    res <- parallel::mclapply(
      seeds,
      function(s) {
        p <- mapear_hparams(p_base)
        p$seed        <- s
        p$num_threads <- 1L  # clave si usás varios cores arriba

        mdl <- lightgbm::lgb.train(
          data    = dtrain,
          valids  = list(valid = dvalidate),
          eval    = "auc",
          param   = p,
          verbose = -100
        )
        auc <- mdl$record_evals$valid$auc$eval[[mdl$best_iter]]
        rm(mdl); gc(FALSE)
        auc
      },
      mc.cores = min(length(seeds), max(1L, parallel::detectCores() - 1L))
    )

    aucs <- unlist(res, use.names = FALSE)

    # agregación robusta
    agg <- match.arg(PARAM$bo$agg, c("mean","median","p25","p75"))
    AUC <- switch(
      agg,
      mean   = mean(aucs, na.rm = TRUE),
      median = median(aucs, na.rm = TRUE),
      p25    = as.numeric(quantile(aucs, 0.25, na.rm = TRUE)),
      p75    = as.numeric(quantile(aucs, 0.75, na.rm = TRUE))
    )
    return(AUC)   # <- devuelve UN escalar, lo que mlrMBO espera

  } else {
    # modo rápido: una sola semilla (la primigenia)
    p <- mapear_hparams(p_base)
    p$seed        <- as.integer(PARAM$semilla_primigenia)
    p$num_threads <- 1L

    mdl <- lightgbm::lgb.train(
      data    = dtrain,
      valids  = list(valid = dvalidate),
      eval    = "auc",
      param   = p,
      verbose = -100
    )
    AUC <- mdl$record_evals$valid$auc$eval[[mdl$best_iter]]
    rm(mdl); gc(FALSE)
    return(AUC)
  }
}
```

seteo de la Bayesian Optimization (complejo)
<br> copiado y pegado de la documentación de la librería

```r
configureMlr(show.learner.output = FALSE)

# configuro la busqueda bayesiana,  los hiperparametros que se van a optimizar
# por favor, no desesperarse por lo complejo
obj.fun <- makeSingleObjectiveFunction(
    fn= EstimarGanancia_AUC_lightgbm, # la funcion que voy a maximizar
    minimize= FALSE, # estoy Maximizando AUC
    noisy= FALSE,
    par.set= PARAM$hipeparametertuning$hs,
    has.simple.signature= FALSE # paso los parametros en una lista
)

# cada 600 segundos guardo el resultado intermedio
ctrl <- makeMBOControl(
    save.on.disk.at.time= 600,
    save.file.path= "HT.RDATA"
)

# indico la cantidad de iteraciones que va a tener la Bayesian Optimization
ctrl <- setMBOControlTermination(
    ctrl,
    iters= PARAM$hipeparametertuning$num_interations  # cantidad de iteraciones inteligentes
)

# defino el método estandar para la creacion de los puntos iniciales
#   los "No Inteligentes"
ctrl <- setMBOControlInfill(ctrl, crit = makeMBOInfillCritEI())

# mas configuraciones
surr.km <- makeLearner(
    "regr.km",
    predict.type= "se",
    covtype= "matern3_2",
    control= list(trace = TRUE)
)
```

```r
# genero archivos con los  "envios" mejores
# suba TODOS los archivos a Kaggle

PARAM$kaggle$competencia <- "data-mining-analista-sr-2025-a"
PARAM$kaggle$cortes <- seq(10000, 12000, by = 500)

# ordeno por probabilidad descendente
setorder(tb_prediccion, -prob)

dir.create("kaggle")

for (envios in PARAM$kaggle$cortes) {

  tb_prediccion[, Predicted := 0L] # seteo inicial a 0
  tb_prediccion[1:envios, Predicted := 1L] # marclo los primeros

  archivo_kaggle <- paste0(
    "./kaggle/KA",
    PARAM$experimento, "_",
    "s", length(PARAM$FT$semillas), "_",
    envios, ".csv")

  # grabo el archivo
  fwrite(tb_prediccion[, list(numero_de_cliente, Predicted)],
    file= archivo_kaggle,
    sep= ","
  )

  # subida a Kaggle, armo la linea de comando
  comando <- "kaggle competitions submit"
  competencia <- paste("-c", PARAM$kaggle$competencia)
  arch <- paste( "-f", archivo_kaggle)

  mensaje <- paste0("-m 'envios=", envios,
  "  semilla=", PARAM$semilla_primigenia,
    "'" )

  linea <- paste( comando, competencia, arch, mensaje)

  salida <- system(linea, intern=TRUE) # el submit a Kaggle
  cat(salida, "\n")
}
```

```r
# grabo los parametros
if( !require("yaml")) install.packages("yaml")
require("yaml")

write_yaml( PARAM, file="PARAM.yml")
```
